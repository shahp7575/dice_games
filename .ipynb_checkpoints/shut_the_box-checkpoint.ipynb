```python
from time import sleep
from random import randint

def game_rules():
    rules = '''
        Welcome to Shut The Box game!!!!!
        
        Here is how it is played:
            1) This is a single-user game
            2) Numbers 1 to 9 will be displayed on the screen - numbered levers
            3) You will get a chance to roll two dices
            4) The number of dots('pips) on the dices will be added
            5) Using that sum you can knock down or close any combination of numbered levers that add up to the sum
            6) For example : if a 2 and a 5 are rolled, the player may choose to close:
                7,   1 & 6,   2 & 5,   3 & 4,   1 & 2 & 4
            7) Note: Until the numbered levers 7-9 are open the player must roll both dices. If these levers are closed then 
                    player may choose to roll one dice or both dice.
            8) You will continue to roll the dice until you can no longer close an appropriate numbered lever(s)
            9) The player's final score will be the the sum of the numbers left on the lever
            10) Best Score - 0 ! And that's when you call Shut the Box!!!!!!!
            
            LET'S DO THIS!!
    '''
    print(rules)

def play_game():
    ready = input("Ready to play(y/n)? ")
    return ready

def roll_dice(how_many):
    
    roll_choice = input("Ready to roll (y/n): ")
    print("Rolling Dice...")
    sleep(1)
    if how_many == 1:
        rolldice1 = randint(1, 6)
        print("Dice 1: %d" %(rolldice1))
        total = rolldice1
    elif how_many == 2:
        rolldice1 = randint(1, 6)
        rolldice2 = randint(1, 6)
        print("Dice 1: %d" %(rolldice1))
        print("Dice 2: %d" %(rolldice2))
        total = rolldice1 + rolldice2
    return total

def how_many_dice(lever_list):
    if max(lever_list) <= 6:
        user_roll = input("How many dice you would like to roll(1 or 2): ")
        total = roll_dice(int(user_roll))
    else:
        total = roll_dice(2)
    return total

def lost():
    print("Sorry you lost :(")
    
def won():
    print("YAYYY you won!!!! SHUT THE BOX!")

def shut_the_box():
    lever = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    total = 0
    cont = True
    lose_streak = 0
    lever_open = []
    while(cont):
        print("Your lever: " + str(lever))
        total = how_many_dice(lever)     
        try_again = True
        while(try_again):
            print("Items in lever: " + str(lever))
            lever_close = input("Which lever you would like to close? (if there is more than one, then enter each with a space): ").split()
            lever_close = list(map(int, lever_close))
            if (sum(lever_close) == total):
                lever = [x for x in lever if x not in lever_close]
                if lever == lever_open:
                    if lose_streak < 3:
                        print("Item entered is not present in the lever. Try again. You can roll %d more times" %(3 - lose_streak))
                        total = how_many_dice(lever)
                        print("Your lever now is: " +str(lever) )
                        lose_streak += 1
                    else:
                        try_again = False
                        cont = False
                        lost()
                else:
                    if lever == []:
                        try_again = False
                        cont = False
                        won()
                    else:
                        print("Your lever now is: " +str(lever) )
                        try_again = False
            elif lose_streak < 3:
                print("Oops looks like you are stuck. Let's roll again. You can roll %d more times" %(3 - lose_streak))
                total = how_many_dice(lever)
                lose_streak += 1
            else:
                try_again = False
                cont = False
                lost()
        lever_open = lever

play = True
while(play):
    game_rules()
    game = play_game()
    
    if game.lower() == 'y':
        shut_the_box()
    break
```
